```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# --- Step 2: Load Dataset ---
file_path = r"C:/Users/Aditya k/OneDrive/Desktop/Customer Demographics Descriptive Analysis/Data/KPMG_Customer_Demography.csv"  # change path as needed
df = pd.read_csv(file_path)
```

```python
print("Shape:", df.shape)
print(df.head())
print(df.info())
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df = df.drop_duplicates()
df = df.dropna(subset=['age', 'wealth_segment', 'owns_car', 'gender'])
```

```python
desc_stats = df.describe(include='all')
print("\nðŸ“Š DESCRIPTIVE STATISTICS:")
print(desc_stats)
```

```python
age_groups = pd.cut(df['age'], bins=[0, 25, 35, 45, 60, 100],labels=['18â€“25', '26â€“35', '36â€“45', '46â€“60', '60+'])
df['age_group'] = age_groups

summary = df.groupby(['age_group', 'wealth_segment'])['past_3_years_bike_related_purchases'].agg(['mean', 'sum', 'count']).reset_index()
print("\nðŸ“ˆ AGE-GROUP & INCOME SUMMARY:")
print(summary.head())
```

```python
out_path = r"C:/Users/Aditya k/OneDrive/Desktop/Customer Demographics Descriptive Analysis/Data/customer_summary.csv"
df.to_csv(out_path, index=False)
print("âœ… Cleaned dataset saved for Power BI import at:", out_path)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='wealth_segment', hue='gender')
plt.title("Gender vs Wealth Segment")
plt.xticks(rotation=45)
plt.show()
```
